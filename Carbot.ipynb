## Data Cleaning

1. Year has many non-year values and change the datatype
2. Price has rows with ASK for PRICE and also change the datatype
3. Kms_driven has km at the end with nan values and datatype issue
4. fuel type has nan values
5. handle the name inconsistency

```python
df=df[df['year'].str.isnumeric()]
```

```python
df
```

```python
df['year']=df['year'].astype(int)
```

```python
df.info()
```

```python
df=df[df['Price']!= "Ask For Price"]
```

```python
df['Price'].unique()
```

```python
df.info()
```

```python
df['Price']=df['Price'].str.replace(',', '').astype(int)
```

```python
df.info()
```

```python
df['kms_driven']= df['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
```

```python
df=df[df['kms_driven'].str.isnumeric()]
```

```python
df['kms_driven']=df['kms_driven'].astype(int)
```

```python
df.info()
```

```python
df=df[~df['fuel_type'].isna()]
```

```python
df['name']=df['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
df['name']
```

```python
df=df.reset_index(drop=True)
```

```python
df
```

```python
df.describe()
```

```python
df=df[df['Price']<6e6].reset_index(drop=True)
```

# Model

```python
x= df.drop(columns= "Price")
y= df['Price']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score
```

```python
ohe= OneHotEncoder()
ohe.fit(x[['name', 'company', 'fuel_type']])
```

```python
column_trans= make_column_transformer((OneHotEncoder(categories= ohe.categories_), ['name', 'company', 'fuel_type']), remainder= "passthrough")
```

```python
lr= LinearRegression()
```

```python
pipe= make_pipeline(column_trans, lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred= pipe.predict(x_test)
```

```python
r2_score(y_test, y_pred)
```

```python
scores=[]
for i in range(1000):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state= i)
    lr= LinearRegression()
    pipe= make_pipeline(column_trans, lr)
    pipe.fit(x_train, y_train)
    y_pred= pipe.predict(x_test)
    scores.append(r2_score(y_test, y_pred))   
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state= np.argmax(scores))
lr= LinearRegression()
pipe= make_pipeline(column_trans, lr)
pipe.fit(x_train, y_train)
y_pred= pipe.predict(x_test)
r2_score(y_test, y_pred)
```
